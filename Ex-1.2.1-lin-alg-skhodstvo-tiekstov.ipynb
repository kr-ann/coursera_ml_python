```python
file_obj = open ('sentences.txt')
```

```python
strings_list = file_obj.readlines()
```

```python
lower_str = []
```

```python
for line in strings_list:
    lower_str.append(line.strip().lower())
```

```python
file_obj.close()
```

```python
splitted = []
```

```python
import re
```

```python
for line in lower_str:
    splitted.append(re.split('[^a-z]', line))
```

```python
for each_list in splitted:
    i = 0
    for word in each_list:
        if (word == '') or (word == '\t') or (word == '\n') or (word=='\r') or (word == '\r\n'):
            del each_list[i]
        i += 1
```

```python
for each_list in splitted:
    for word in each_list:
        print word
```

```python
i = 0
wf = dict()
for each_list in splitted:
    for word in each_list:
        if not (word in wf):
            wf[word] = i
            i += 1
```

```python
import numpy as np
```

```python
c = np.zeros((len(splitted), len(wf)))
```

```python
i=0 # a sentence 
for each_list in splitted:
    for word in each_list:
        c[i,wf[word]]+=1
    i += 1   
```

```python
from scipy.spatial.distance import cosine
```

```python
res = []
i = 1
while (i != len(splitted)):
    res.append(cosine(c[0,:].reshape(1,len(wf)), c[i,:].reshape(1,len(wf))))
    i += 1
```

```python
r1 = 1
r2 = 1
i = 0
j = 0
for l in res:
    if r1>l:
        r1 = l
for l in res:
    if (r2>l) and (l != r1):
        r2 = l
while res[i] != r1:
    i+=1
while res[j] != r2:
    j+=1
print 'первое значение - ', r1, 'номер предложения - ', i + 1
print '\nвторое значение - ', r2, 'номер предложения - ', j + 1
print res
```
